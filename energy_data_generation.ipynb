```r
%python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Definindo uma seed para garantir a reprodutibilidade dos dados gerados
np.random.seed(42)
random.seed(42)

print("Iniciando a geração dos dados para a demo...")

# ==========================================
# 1. Tabela Plant (Plantas de Geração)
# ==========================================
plants = {
    'plant_id': [1, 2, 3, 4, 5],
    'plant_name': ['Usina Solar Alpha', 'Parque Eólico Beta', 'Hidrelétrica Gamma', 'Usina Solar Delta', 'Parque Eólico Epsilon'],
    'plant_type': ['Solar', 'Wind', 'Hydro', 'Solar', 'Wind'],
    'capacity_mw': [50, 120, 300, 80, 150], # Capacidade em Megawatts
    'location': ['Região Norte', 'Região Sul', 'Região Centro-Oeste', 'Região Nordeste', 'Região Sul'],
    'status': ['Ativa', 'Ativa', 'Ativa', 'Em Manutenção', 'Ativa']
}
df_plant = pd.DataFrame(plants)
df_plant.to_csv('./plant.csv', index=False)
print("-> plant.csv gerado com sucesso!")

# ==========================================
# 2. Tabela Histórico de Consumo/Produção (100 linhas)
# ==========================================
start_date = datetime(2023, 1, 1)
# Gerando 100 dias aleatórios dentro de um intervalo de 30 dias (haverá dias repetidos para plantas diferentes)
dates = [start_date + timedelta(days=random.randint(0, 30)) for _ in range(100)]
plant_ids_consumo = [random.choice([1, 2, 3, 4, 5]) for _ in range(100)]

producao_mwh = []
consumo_interno_mwh = []

for pid in plant_ids_consumo:
    # Simulação baseada na capacidade da planta
    base_cap = df_plant[df_plant['plant_id'] == pid]['capacity_mw'].values[0]
    
    # Produção diária simulada (Fator de capacidade entre 10% e 90% * capacidade * 24h)
    producao_mwh.append(round(np.random.uniform(0.1, 0.9) * base_cap * 24, 2))
    
    # Consumo diário (Energia gasta internamente pela planta para operar)
    consumo_interno_mwh.append(round(np.random.uniform(0.01, 0.05) * base_cap * 24, 2))

df_consumo = pd.DataFrame({
    'record_id': range(1, 101),
    'date': [d.strftime('%Y-%m-%d') for d in dates],
    'plant_id': plant_ids_consumo,
    'energy_produced_mwh': producao_mwh,
    'internal_consumption_mwh': consumo_interno_mwh # Consumo diário
})

# Ordenando por data para ficar mais realista e reindexando
df_consumo = df_consumo.sort_values(by=['date', 'plant_id']).reset_index(drop=True)
df_consumo['record_id'] = range(1, 101) # Refazendo o ID
df_consumo.to_csv('historico_consumo.csv', index=False)
print("-> historico_consumo.csv gerado com sucesso (100 linhas)!")

# ==========================================
# 3. Tabela Histórico de Manutenção (30 linhas)
# ==========================================
reasons = [
    'Troca de Inversor', 
    'Limpeza de Painéis', 
    'Falha na Turbina', 
    'Inspeção Anual Preventiva', 
    'Problema no Gerador', 
    'Reparo de Cabeamento', 
    'Atualização de Software SCADA'
]

# Distribuição de datas ao longo de 6 meses
dates_manutencao = [start_date + timedelta(days=random.randint(0, 180)) for _ in range(30)]

df_manutencao = pd.DataFrame({
    'maintenance_id': range(1, 31),
    'plant_id': [random.choice([1, 2, 3, 4, 5]) for _ in range(30)],
    'date': [d.strftime('%Y-%m-%d') for d in dates_manutencao],
    'maintenance_reason': [random.choice(reasons) for _ in range(30)],
    'duration_hours': [random.randint(2, 72) for _ in range(30)], # Tempo que a planta ficou parada
    'cost_usd': [round(np.random.uniform(500, 15000), 2) for _ in range(30)] # Custo da manutenção
})

# Ordenando por data
df_manutencao = df_manutencao.sort_values(by='date').reset_index(drop=True)
df_manutencao.to_csv('./historico_manutencao.csv', index=False)
df_manutencao['maintenance_id'] = range(1, 31)
df_manutencao.to_csv('./historico_manutencao.csv', index=False)
print("-> historico_manutencao.csv gerado com sucesso (30 linhas)!")

print("Todos os arquivos foram gerados no seu diretório atual.")
```
